```python
# LIBRARY

import numpy as np # vector manipulation

# THIS IS FOR PLOTTING
%matplotlib inline
import matplotlib.pyplot as plt # side-stepping mpl backend
import warnings
warnings.filterwarnings("ignore")
from scipy.stats import norm
```

```python
mu_0=40
c=np.arange(0,100,0.1)
fig = plt.figure(figsize=(5,5))

plt.plot(c,mu_0+mu_0/100*c,'-',color='blue',label="Right")
plt.plot(c,mu_0-mu_0/100*c,'-',color='red',label="Left")
plt.grid()
plt.xlabel('Coherence')
plt.ylabel('Firing Rate (Hz)')
#plt.xlim((-100,1000))
#plt.ylim((0,Threshold))
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

# plt.
#plt.ylim((30,50))
plt.show()

```

```python
Stimulus_intensity=np.arange(10,70,1)
plt.plot(Stimulus_intensity, norm.pdf(Stimulus_intensity,mu_0+mu_0/100*6.4,10))
plt.plot(Stimulus_intensity, norm.pdf(Stimulus_intensity,mu_0-mu_0/100*6.4,10),'r')
plt.xlabel('Stimulus Intensity')
plt.show()
```

```python
R=np.random.normal(mu_0+mu_0/100*6.4,10,20)
L=np.random.normal(mu_0-mu_0/100*6.4,10,20)
time=np.arange(0,1001,50)
R=np.concatenate((R,np.zeros(1)))
R=np.concatenate((np.zeros(1),R))
L=np.concatenate((L,np.zeros(1)))
L=np.concatenate((np.zeros(1),L))
```

```python
Stimulus_intensity=np.arange(10,70,1)
for i in range(0,len(R)-2):
    plt.hlines(y=R[i+1], xmin=time[i], xmax=time[i+1], linewidth=2, color='b')
    plt.hlines(y=L[i+1], xmin=time[i], xmax=time[i+1], linewidth=2, color='r')
    plt.vlines(ymin=R[i], ymax=R[i+1], x=time[i], linewidth=2, color='b')
    plt.vlines(ymin=L[i], ymax=L[i+1], x=time[i], linewidth=2, color='r')
      
plt.hlines(y=0, xmin=-1000, xmax=0, linewidth=2, color='r')
plt.hlines(y=0, xmin=1000, xmax=2000, linewidth=2, color='r')
plt.vlines(ymin=R[-1], ymax=R[-2], x=time[-1], linewidth=2, color='b')
plt.vlines(ymin=L[-1], ymax=L[-2], x=time[-1], linewidth=2, color='r')
 



plt.xlabel('time(ms)')
plt.ylabel('Sample Stimulus (Hz)')

plt.xlabel('time (ms)')
plt.show()

```
