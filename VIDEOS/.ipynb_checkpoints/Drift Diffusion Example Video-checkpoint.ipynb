## Drift Diffusion Video


A simple example of plotting the drift diffusion as a video


```python
## Library
import numpy as np
import math 
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

import matplotlib.gridspec as gridspec # subplots
import warnings
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 14}

plt.rc('font', **font)
from celluloid import Camera
```

```python
N=50
x=1.0*np.ones(N+1)
T=1000
```

```python
steps = np.random.normal(0.1, 1, size=(N, T))

# compute the cumulative sum of all the steps over the time axis
sim = np.cumsum(steps, axis=1)


```

```python
with plt.xkcd():

    fig = plt.figure(figsize=(8, 6))
    camera=Camera(fig)
    ax = fig.add_subplot(1,1,1)
    for n in range(0,1000+1,50): 
        ax.plot(sim[0,0:n],color='black')#,label='Hares')
        ax.plot(0,0,'o',color='blue')#,label='Hares')
        ax.text(0,85,'Decision Bound' ,color='green')
        ax.text(0,-95,'Decision Bound' ,color='green')
        ax.text(-100,-20,'Start Point' ,color='blue')

        ax.axhline(100,color='green')#,label='Hares')
        ax.axhline(-100,color='green')#,label='Hares')
        ax.axhline(0,linestyle='dashed',color='grey')#,label='Hares')

        ax.set_ylabel('Evidence')
        ax.set_xlabel('Time (ms)')
        if n >400: 
            ax.arrow(300, sim[0,300]+1, 80, (sim[0,400]-sim[0,300])*1.4, head_width=10, head_length=10,color='blue')
            ax.text(300,sim[0,300]+10,'Drift',rotation=40, color='blue')
        camera.snap()
        ax.set_ylim(-150,150)
        ax.set_xlim(-150,1200)
    animation = camera.animate()
    animation.save('DDM.gif', writer='Pillow', fps=5)
```
